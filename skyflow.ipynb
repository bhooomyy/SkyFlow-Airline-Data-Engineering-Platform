```python
#Transform
dim_airline=airline_data[[
    'DOT_ID_Reporting_Airline',
    'Reporting_Airline',
    'IATA_CODE_Reporting_Airline']].drop_duplicates()
dim_airline.columns=['airline_id','airline_code','iata_code']
print(dim_airline.head())
print(dim_airline.shape)
print(dim_airline.info())
```

```python
origin_airport=airline_data[[
    'OriginAirportID',
    'Origin',
    'OriginCityName',
    'OriginState',
    'OriginStateName']].drop_duplicates()
origin_airport.columns=['airport_id','airport_code','city_name','state_code','state_name']

dest_airport=airline_data[[
    'DestAirportID',
    'Dest',
    'DestCityName',
    'DestState',
    'DestStateName']].drop_duplicates()
dest_airport.columns=['airport_id','airport_code','city_name','state_code','state_name']

dim_airport=pd.concat([origin_airport,dest_airport]).drop_duplicates()

print("dim_airport preview: ",dim_airport.head())
print(dim_airport.shape)
print(dim_airport.info())
```

```python
dim_date=airline_data[[
    'FlightDate',
    'Year',
    'Quarter',
    'Month',
    'DayofMonth',
    'DayOfWeek'
]].drop_duplicates()

dim_date=dim_date.sort_values('FlightDate').reset_index(drop=True)

dim_date.columns=[
    'flight_date',
    'yearr',
    'quarterr',
    'monthh',
    'day_of_month',
    'day_of_week'
]

print("dim_date preview:",dim_date.head())
print("dim_date shape:",dim_date.shape)
print(dim_date.info())
```

```python
fact_flights=airline_data[[
    'FlightDate',
    'DOT_ID_Reporting_Airline',
    'Flight_Number_Reporting_Airline',
    'Tail_Number',
    'OriginAirportID',
    'DestAirportID',
    'CRSDepTime',
    'DepTime',
    'DepDelay',
    'DepDelayMinutes',
    'CRSArrTime',
    'ArrTime',
    'ArrDelay',
    'ArrDelayMinutes',
    'Cancelled',
    'Diverted',
    'TaxiOut',
    'TaxiIn',
    'AirTime',
    'Distance',
    'CarrierDelay',
    'WeatherDelay',
    'NASDelay',
    'SecurityDelay',
    'LateAircraftDelay'
]].copy()

fact_flights.columns=[
    'flight_date',
    'airline_id',
    'flight_number',
    'tail_number',
    'origin_airport_id',
    'dest_airport_id',
    'scheduled_departure_time',
    'actual_departure_time',
    'departure_delay',
    'departure_delay_minutes',
    'scheduled_arrival_time',
    'actual_arrival_time',
    'arrival_delay',
    'arrival_delay_minutes',
    'cancelled',
    'diverted',
    'taxi_out',
    'taxi_in',
    'air_time',
    'distance',
    'carrier_delay',
    'weather_delay',
    'nas_delay',
    'security_delay',
    'late_aircraft_delay'
]

print("fact_flights preview:",fact_flights.head())
print("\nfact_flights shape:",fact_flights.shape)
print(fact_flights.info())
```

```python
#Data quality validation
print("Null values in fact_flights:\n",fact_flights.isnull().sum())
```

```python
print("\nCheck cancelled flights:",fact_flights['cancelled'].value_counts())
print("\nCheck diverted flights:",fact_flights['diverted'].value_counts())
```

```python
fact_flights['tail_number']=fact_flights['tail_number'].fillna('UNKNOWN')
delay_columns=[
    'carrier_delay',
    'weather_delay',
    'nas_delay',
    'security_delay',
    'late_aircraft_delay'
]
fact_flights[delay_columns]=fact_flights[delay_columns].fillna(0)

print(fact_flights.isnull().sum())
```
